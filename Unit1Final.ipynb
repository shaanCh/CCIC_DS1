```python
import tracemalloc
import random

#The Algorithm
def gnomeSort( arr, n):
    index = 0
    while index < n:
        if index == 0:
            index = index + 1
        if arr[index] >= arr[index - 1]:
            index = index + 1
        else:
            arr[index], arr[index-1] = arr[index-1], arr[index]
            index = index - 1
  
    return arr
  
#makes a list of random integers from -100 to 100
arr = []
list_length = 8
for i in range(0,list_length//2):
    x = random.randint(-100000, 0)
    arr.append(x)
for i in range(0,list_length//2):
    y = random.randint(0, 100000)
    arr.append(y)
print(arr)
n = len(arr)

#Runs Algorithm and times execution
tracemalloc.start()
x1, y1 = tracemalloc.get_traced_memory()
arr = gnomeSort(arr, n)
x2, y2 = tracemalloc.get_traced_memory()
tracemalloc.stop()
print ("Sorted sequence after applying Gnome Sort :")
print(arr)
print("y1:" + str(y1))
print("y2:" + str(y2))
print(y2-y1)
```
